## Draw Line

```
import matplotlib.pyplot as plt
import numpy as np
t=np.linspace(0,2*np.pi,50)
x=np.sin(t)
y=np.cos(t)

plt.figure()
#plt.plot(x)
plt.plot(y)
plt.title("Cos")
```

#Line Plot

```
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,2*np.pi,50)
y1=np.sin(x)
y2=np.sin(2*x)
plt.figure(); #create figure
#plt.plot(y1)
#plt.plot(x,y1)
#plt.plot(x,y1,'y') #'y': yellow
plt.plot(x,y1,'r-o') # red market only circle

plt.clf()
plt.plot(x,y1,"g-o",x,y2,"b-*")
plt.legend(["sin(x)", "sin(2x)"])

```

#Scatter Plot

```
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import rand
N=50
x=np.linspace(0,10,N)
#print(x)
e=rand(N)*5.0 #noise
#print(e)
y1=x+e
areas=rand(N)*100
#plt.scatter(x,y1,s=areas)
colors=rand(N)
plt.scatter(x,y1,s=areas,c=colors)
plt.colorbar()
plt.title("Random scatter")
```

#Image Plot Correlation Plot

```
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import rand
#create some random data
e1=rand(100)
e2=rand(100)*2
e3=rand(100)*50
e4=rand(100)*100

#correlation matrix
#https://www.displayr.com/what-is-a-correlation-matrix/
corrMatrix=np.corrcoef([e1,e2,e3,e4])
print(corrMatrix)
#plot corr matrix as image

plt.imshow(corrMatrix,cmap='GnBu')
#cmaps['Sequential'] = [
#            'Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds',
#            'YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu',
#            'GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn']
plt.colorbar()
```

#Multiple plot and subplot

```
import matplotlib.pyplot as plt
import numpy as np
t=np.linspace(0,2*np.pi)
x=np.sin(t)
y=np.cos(t)
# To divide the plotting area
plt.subplot(2,1,1)
plt.plot(x,label="sin")
plt.legend()
# activate a new plot
plt.subplot(2,1,2)
plt.plot(y,label="cos")
plt.legend()
```

# Histogram plot

```
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import randint
data=randint(10000, size=(10,10000))
# Approx norm distribution
x=np.sum(data, axis=0)
#set up for stacked plots
plt.subplot(2,1,1)
plt.hist(x,color='g')
#plot cumulative dist
plt.subplot(2,1,2)
plt.hist(x,cumulative=True)
```
