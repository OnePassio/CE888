# Linear Regression
Let's fabricate some data that shows a roughly linear relationship between page speed and amount purchased:

```
%matplotlib inline
import numpy as np
from pylab import *

pageSpeeds=np.random.normal(3.0,1.0,1000)
purchaseAmount=100-(pageSpeeds+np.random.normal(0,0.1,1000))*3

scatter(pageSpeeds,purchaseAmount)
```

As we only have 2 features, we can keep it simple and just use scipy.state.linregress:

```
from scipy import stats
slope, intercept, r_value, p_value, std_error= stats.linregress(pageSpeeds, purchaseAmount)

print("R-Squared: {}".format(r_value**2))
```

Let's use the slope and intercept we got from the regression to plot predicted values vs observed:

```
import matplotlib.pyplot as plt
def predict(x):
  return slope*x+intercept

fitLine=predict(pageSpeeds)
plt.scatter(pageSpeeds,purchaseAmount)
plt.plot(pageSpeeds,fitLine,c='r')
plt.show
```

# Polynomial Regression

```
import matplotlib.pyplot as plt
from pylab import *
import numpy as np
np.random.seed(2) # keep random persistent
pageSpeeds=np.random.normal(3,1,1000)
purchaseAmount=np.random.normal(50,10,1000)/pageSpeeds

plt.scatter(pageSpeeds,purchaseAmount)
```

```
x=np.array(pageSpeeds)
y=np.array(purchaseAmount)

p4=np.poly1d(np.polyfit(x,y,4))
print(p4)
xp=np.linspace(0,7,100)
plt.scatter(x,y)
plt.plot(xp,p4(xp),c='r')
plt.show()
```

Check R-Squared

```
from sklearn.metrics import r2_score
r2=r2_score(y,p4(x))
print("R Squared is {}".format(r2))
```
